 Задание 3
 Напишите функцию, которая возвращает название валюты (поле ‘Name’) с максимальным значением курса с помощью сервиса https://www.cbr-xml-daily.ru/daily_json.js

```python
import requests

def find_max_valute(valutes):
    m = max(valutes.items(), key = lambda x : x[1]['Value'])
    return (m[0], m[1]['Name'], m[1]['Value'])

def get_valute_name_with_max_value() :
    service_url = 'https://www.cbr-xml-daily.ru/daily_json.js'
    response = requests.get(service_url)
    valutes = response.json()['Valute']
    return find_max_valute(valutes)

print(get_valute_name_with_max_value())
```

 Задание 4
 Последнее упражнение с занятия
 Добавьте в класс еще один формат, который возвращает название валюты (например, ‘Евро’).
 Добавьте в класс параметр diff (со значениями True или False), который в случае значения True в методах eur и usd будет возвращать не курс валюты, а изменение по сравнению в прошлым значением.

```python
class Rate:
    def __init__(self, format='value', diff=False):
        self.format = format
        self.diff = diff
    
    def exchange_rates(self):
        """
        Возвращает ответ сервиса с информацией о валютах в виде:
        
        {
            'AMD': {
                'CharCode': 'AMD',
                'ID': 'R01060',
                'Name': 'Армянских драмов',
                'Nominal': 100,
                'NumCode': '051',
                'Previous': 14.103,
                'Value': 14.0879
                },
            ...
        }
        """
        r = requests.get('https://www.cbr-xml-daily.ru/daily_json.js')
        return r.json()['Valute']
    
    def make_format(self, currency):
        """
        Возвращает информацию о валюте currency в двух вариантах:
        - полная информация о валюте при self.format = 'full':
        Rate('full').make_format('EUR')
        {
            'CharCode': 'EUR',
            'ID': 'R01239',
            'Name': 'Евро',
            'Nominal': 1,
            'NumCode': '978',
            'Previous': 79.6765,
            'Value': 79.4966
        }
        
        Rate('value').make_format('EUR')
        79.4966
        """
        response = self.exchange_rates()
        
        if currency in response:
            if self.format == 'full':
                return response[currency]
            
            if self.format == 'value':
                if self.diff :
                    return response[currency]['Value'] - response[currency]['Previous'] 
                return response[currency]['Value']

            if self.format == 'name':
                return response[currency]['Name']
        
        return 'Error'
    
    def eur(self):
        """Возвращает курс евро на сегодня в формате self.format"""
        return self.make_format('EUR')
    
    def usd(self):
        """Возвращает курс доллара на сегодня в формате self.format"""
        return self.make_format('USD')

# использование
curr = Rate()
print(curr.usd())
print(curr.eur())

curr = Rate('full')
print(curr.usd())
print(curr.eur())

curr = Rate('name')
print(curr.usd())
print(curr.eur())

curr = Rate(diff = True)
print(curr.usd())
print(curr.eur())

```

 Задание 5
 Напишите функцию, возвращающую сумму первых n чисел Фибоначчи

```python
def calc_sum_fibbon(n):
    if n == 0 :
        return 0;
    elif n == 1 :
        return 1;
    elif n == 2:
        return 2;

    a, b, sum = 1, 1, 2
    n -= 2
    while n > 0 :
        next = a + b
        sum += next
        a = b
        b = next
        n -= 1
    
    return sum

for n in [0, 1, 2, 4, 6, 8, 65] :
    sum = calc_sum_fibbon(n)
    print("sum of {0} fibbonachi numbers = {1}".format(n, sum))
```

 Задание 6
 Напишите функцию, преобразующую произвольный список вида [‘2018-01-01’, ‘yandex’, ‘cpc’, 100] в словарь {‘2018-01-01’: {‘yandex’: {‘cpc’: 100}}}

```python
def make_dict(lst) :
    if len(lst) < 2 : 
        raise Exception('список должен иметь хотя бы 2 элемента')        
    if len(lst) == 2 :
        return {lst[0] : lst[1]}
    
    return {lst[0] : make_dict(lst[1:])}

input_list = ['2018-01-01', 'yandex', 'cpc', 100]
output_dict = make_dict(input_list)
print(input_list)
print(output_dict)
```
