*Задание 3*
 Список поисковых запросов. Получить распределение количества слов в них. Т. е. поисковых запросов из одного слова 5%, из двух - 7%, из трех - 3% итд.

```python
queries = [
   'смотреть сериалы онлайн',
   'новости спорта',
   'афиша кино',
   'курс доллара',
   'сериалы этим летом',
   'курс по питону',
   'сериалы про спорт',
]

counts = [len(x.split(' ')) for x in queries]
distribution_count = {}
for item in counts:
    if item in distribution_count:
        distribution_count[item] += 1
    else :
        distribution_count[item] = 1
distribution_percent = {}
for words_count, queries_count in distribution_count.items():
    distribution_percent[words_count] = queries_count * 100 / len(queries)

for words_count, percent in distribution_percent.items() :
    print("Quieries with {0} words -> distribution {1:.2f}%".format(words_count, percent))
```

 *Задание 5*
 Дан поток логов по количеству просмотренных страниц для каждого пользователя. Список отсортирован по ID пользователя. Вам необходимо написать алгоритм, который считает среднее значение просмотров на пользователя. Т. е. надо посчитать отношение суммы всех просмотров к количеству уникальных пользователей. Учтите, что весь список stream не помещается в оперативную память, т. е. его нужно обрабатывать поэлементно в цикле.

```python
stream = [
    '2018-01-01,user1,3',
    '2018-01-07,user1,4',
    '2018-03-29,user1,1',
    '2018-04-04,user1,13',
    '2018-01-05,user2,7',
    '2018-06-14,user3,4',
    '2018-07-02,user3,10',
    '2018-03-21,user4,19',
    '2018-03-22,user4,4',
    '2018-04-22,user4,8',
    '2018-05-03,user4,9',
    '2018-05-11,user4,11',
]

prev_user = ''
users_count = 0
views_count = 0
for user_str in stream :
   parts = user_str.split(',')
   if prev_user != parts[1] :
      prev_user = parts[1]
      users_count += 1
   views_count += int(parts[2])

views_per_user = views_count / users_count
print("Views per user = {0:.2f}".format(views_per_user))
```

 *Задание 6*
 Дана статистика рекламных кампаний по дням. Напишите алгоритм, который по паре дата-кампания ищет значение численного столбца. Т. е. для даты '2018-01-01' и 'google' нужно получить число 25. Считайте, что все комбинации дата-кампания уникальны, а список stats легко помещается в оперативной памяти.

```python
stats = [
    ['2018-01-01', 'google', 25],
    ['2018-01-01', 'yandex', 65],
    ['2018-01-01', 'market', 89],
    ['2018-01-02', 'google', 574],
    ['2018-01-02', 'yandex', 249],
    ['2018-01-02', 'market', 994],
    ['2018-01-03', 'google', 1843],
    ['2018-01-03', 'yandex', 1327],
    ['2018-01-03', 'market', 1764],
]

date, company = '2018-01-01', 'google'
stat = list(filter(lambda x : x[0] == date and x[1] == company, stats))[0][2]
print(stat)
```
